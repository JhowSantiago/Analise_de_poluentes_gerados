```python
import pandas as pd
```

```python
df = pd.read_csv('./poluentes.csv', sep=';')
```

```python
df['Quantidade'] = pd.to_numeric(df['Quantidade'].str.replace(',', ''), downcast='float')
df.info()
```

```python
df.head(10)
```

```python
cidades = ['SANTOS', 'SAO VICENTE', 'GUARUJA','PRAIA GRANDE', 'CUBATAO']
df_bs = df[df.Município.isin(cidades)]

df_sp = df[df.Município.isin(['SAO PAULO'])]
```

```python
df_bs.groupby(['Ano']).sum()['Quantidade']
```

```python
df_bs.groupby(['Ano', 'Poluente emitido']).sum()['Quantidade']
```

```python
df_bs.groupby(['Categoria de Atividade']).sum()['Quantidade'].sort_values(ascending=False)
```

```python
df_bs.groupby(['Categoria de Atividade', 'Poluente emitido']).sum()['Quantidade'].sort_values(ascending=False)
```

```python
df_bs.groupby(['Município'])['Quantidade'].sum().sort_values(ascending=False)
```

```python
result_df = pd.DataFrame({'Soma': df_bs.groupby(['Município', 'Poluente emitido'])['Quantidade'].sum()})

result_df.sort_values(by=['Município', 'Soma'], ascending=[True, False])
```
